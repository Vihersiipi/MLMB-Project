```python
labels = pd.read_csv("./data/Peptide_Targetlabels.csv", index_col = 0)
labels
```

```python
# Used to find the columns with meaninful variation
for label in labels.drop(["MAIN_Category", "isSinglepurpose"], axis = 1).columns:
    total = labels[label].sum()
    if total > 17265 * 0.01 and total < 17265 * 0.99:
        print(label, labels[label].sum())
```

```python
# Full Code
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import accuracy_score


# Just specifies the target value, this is just used for easy testing of different predictors
target_value = "MAIN_Category"
# useall to drop or not drop the "isSinglepurpose" = FALSE columns in the dataframe
useall = True

# Dropping rows with NA-values (none exist in this dataset)
embeddings_finetuned_fixed = embeddings_finetuned.dropna()

# Selecting only the Rows with "isSinglepurpose" = True
embeddings_finetuned_fixed = embeddings_finetuned_fixed[labels["isSinglepurpose"]] if not useall else embeddings_finetuned_fixed

# Splitting the data into a training and testing sets
training_data, testing_data = train_test_split(embeddings_finetuned_fixed, test_size = 0.20, random_state = 24)

n = 40

# Setting up PCA and a Normal scaler
# Here 40 is selected as it mostly covers the variance of the data set
pca = PCA(n_components = n)
scaler = StandardScaler()

# Normalization to zero mean and 1 variance

# Fits the data at the same time normalizes the training set
normalized_training = scaler.fit_transform(training_data)
# Normalizes the Testing set using the same transformation as the Training set
normalized_testing = scaler.transform(testing_data)

# PCA + Preprocessing
pca_training = pca.fit_transform(normalized_training)
pca_testing = pca.transform(normalized_testing)

# Assigning the components from PCA to a dataset with names "Component #" for each componennt
pca_training = pd.DataFrame(index = training_data.index, data = pca_training, columns = [f"Component {i}" for i in range(1, n + 1)])
pca_testing = pd.DataFrame(index = testing_data.index, data = pca_testing, columns = [f"Component {i}" for i in range(1, n + 1)])

# Joins the label data to the PCA data
pca_training = pca_training.join(labels)
pca_testing = pca_testing.join(labels)

# MODEL - Support Vectors Machine 
svm = SVC(decision_function_shape = 'ovo')
# Fits the SVC-model with the PCA data 
# Here "[f"Component {i}" for i in range(1, n + 1)]" selects just the columns with the PCA components from the dataframe
svm.fit(pca_training[[f"Component {i}" for i in range(1, n + 1)]], pca_training[target_value])

# Prediction using the model we just trained
training = svm.predict(pca_training[[f"Component {i}" for i in range(1, n + 1)]])
testing = svm.predict(pca_testing[[f"Component {i}" for i in range(1, n + 1)]])

# Saves the predicted values of the model for the MAIN category for the
pca_training["PRED_MAIN"] = training
pca_testing["PRED_MAIN"] = testing

cm = confusion_matrix(pca_testing[target_value], testing)
cm_dislpay = ConfusionMatrixDisplay(confusion_matrix = cm)
cm_dislpay.plot()

print(f"Results for predicting {target_value} components")
print("Training accuracy", accuracy_score(pca_training[target_value], pca_training["PRED_MAIN"]))
print("Testing accuracy", accuracy_score(pca_testing[target_value], pca_testing["PRED_MAIN"]))
print()
```

```python
# List contain all columns that have between 0.01 and 0.99 occurancy rate in the column
# ie. The rows that have a meaningful amount of variation
target_labels = [
    "Anti-gram+",
    "Anti-gram-",
    "AntiHIV",
    "Antibacterial",
    "Anticancer",
    "Antimicrobial",
    "Antiparasitic",
    "Antiviral",
    "Cell_Communication",
    "Drug_delivery",
    "Glucose_metabolism",
    "Metabolic_regulatory",
    "Neuropeptide"]

# Iterate trough all the meaningul categories
for target_purpose in target_labels:

    # Same Stuff as last time but for the individual values

    svm = SVC()
    svm.fit(pca_training[[f"Component {i}" for i in range(1, n + 1)]], pca_training[target_purpose])

    training = svm.predict(pca_training[[f"Component {i}" for i in range(1, n + 1)]])
    testing = svm.predict(pca_testing[[f"Component {i}" for i in range(1, n + 1)]])

    pca_training[f"PRED_{target_purpose}"] = training
    pca_testing[f"PRED_{target_purpose}"] = testing

    print(f"Results for predicting {target_purpose} ", end="")
    print(accuracy_score(pca_testing[target_purpose], pca_testing[f"PRED_{target_purpose}"]))
```
